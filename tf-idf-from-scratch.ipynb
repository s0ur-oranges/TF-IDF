```python
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv('../input/bbc-text-clscsv/bbc_text_cls.csv')
idx = 0
word2idx = {}
tokenized_docs = []
for doc in df['text']:
  words = word_tokenize(doc.lower())
  doc_as_int = []
  for word in words:
    if word not in word2idx:
      word2idx[word] = idx
      idx += 1

    doc_as_int.append(word2idx[word])
  tokenized_docs.append(doc_as_int)

idx2word = {v:k for k, v in word2idx.items()}  #reverse mapping
```

```python
N = len(df['text'])
V = len(word2idx)
tf = np.zeros((N, V))

for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    tf[i, j] += 1
```

```python
# compute IDF
document_freq = np.sum(tf > 0, axis=0) # document frequency (shape = (V,))
idf = np.log(N / document_freq)

# compute TF-IDF
tf_idf = tf * idf

```

Now lets test the model for some random document , and print the top 10 terms according to tf-idf

```python
i = np.random.choice(N)
row = df.iloc[i]
print("Label:", row['labels'])
print("Text:", row['text'].split("\n", 1)[0])
print("Top 10 terms:")

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:10]:
  print(idx2word[j])
```
